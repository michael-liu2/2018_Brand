Upload excel file - these values were taken from the results obtained from the actual analysis. The data file contains the chemical name, current UK spending in USD, and projected UK spending in USD.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 

data = pd.read_excel("values.xlsx")
```

We sort the data here by total spending for the purposes of making a clean graph

```python
data = data.sort_values(by='sum_cost_US_USD', ascending=False)
```

```python
data.head(5)
```

We graph data for actual and projected UK spending in USD for each chemical.

```python
ind = np.arange(len(data))
width = 0.4

fig, ax = plt.subplots(figsize = (20, 20))
ax.barh(ind, data['sum_cost_US_USD'], width, color='red', label='2018 Projected English Spending')
ax.barh(ind + width, data['sum_cost_UK_USD'], width, color='green', label='2018 English Spending')

ax.set(yticks= ind + width, yticklabels=data['chemical'], ylim=[2*width - 1, len(data)])
ax.legend()
ax.set_xlabel('Spending, $ (100 million)')


plt.savefig('UK_US_price.pdf')
```
